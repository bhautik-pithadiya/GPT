```python
import pandas as pd
```

```python
data = open('../data/dataset_v1.txt')
```

```python
data = data.readlines()
```

```python
data
```

```python
splitted_Data = [x.split(" ") for x in data]
```

```python
splitted_Data
```

```python
len(splitted_Data)
```

```python
x = " ".join([y for x in splitted_Data for y in x])
```

```python
x = x.split(' ')
```

```python
block_size = 256
```

```python
data_df = pd.DataFrame()
```

```python
i=0
while i < len(x)-block_size:
    data = {
        "X":x[i:i+block_size],
        "y": x[i+1 : i + block_size + 1]
    }
    data_df = pd.concat([data_df,pd.DataFrame(data)])
    i+=1
```
