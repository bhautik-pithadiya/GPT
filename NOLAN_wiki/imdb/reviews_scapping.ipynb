```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd

# Start a WebDriver session
driver = webdriver.Chrome()  # You'll need to download the appropriate WebDriver for your browser

# Load the IMDb page with reviews
driver.get('https://www.imdb.com/title/tt0154506/reviews?ref_=tt_urv')  # Example IMDb page URL

# Create DataFrame to store reviews
interstellar_data = pd.DataFrame(columns=['Review'])

# Track the number of reviews loaded
review_count = 0

# Scroll down the page to load more reviews
while True:
    # Scroll to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Add a short delay to allow content to load
    
    # Extract and store reviews
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, 'html.parser')
    reviews = soup.find_all('div', class_='text show-more__control')
    for review in reviews[review_count:]:
        data = {
            'Review': review.get_text().strip()
        }
        data_df = pd.DataFrame([data])
        interstellar_data = pd.concat([interstellar_data,data_df],ignore_index=True)
    
    # Update the review count
    review_count = len(reviews)
    
    # Check if the "load more" button is no longer present
    load_more_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'ipl-load-more__button'))
    )
    if not load_more_button.is_displayed():
        break

    # Scroll down again if the button is present
    driver.execute_script("arguments[0].scrollIntoView();", load_more_button)
    time.sleep(2)  # Add a short delay to allow content to load

    # Click the "load more" button
    load_more_button.click()

# Close the WebDriver session
driver.quit()

# Save DataFrame to a CSV file
interstellar_data.to_csv('interstellar_reviews.csv', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd

# Start a WebDriver session
driver = webdriver.Chrome()

# Load the IMDb page with reviews
driver.get('https://www.imdb.com/title/tt0154506/reviews?ref_=tt_urv')

# Create list to store reviews
reviews_list = []

# Track the number of reviews loaded
review_count = 0

# Track the number of attempts
attempt = 1

# Set maximum number of attempts
max_attempts = 10

# Set maximum number of reviews to capture
max_reviews = 257

# Main loop to gather reviews
while review_count < max_reviews and attempt <= max_attempts:
    # Scroll to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Add a short delay to allow content to load
    
    # Extract and store reviews
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, 'html.parser')
    reviews = soup.find_all('div', class_='text show-more__control')
    for review in reviews[review_count:]:
        reviews_list.append({'Review': review.get_text().strip()})
        review_count += 1
    
    # Check if the desired number of reviews is reached
    if review_count >= max_reviews:
        break
    
    # Attempt to click the "load more" button
    try:
        load_more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CLASS_NAME, 'ipl-load-more__button'))
        )
        load_more_button.click()
        time.sleep(2)
        attempt = 1  # Reset attempt count if successful
    except:
        attempt += 1  # Increment attempt count if unsuccessful

# Close the WebDriver session
driver.quit()

# Create DataFrame from the list of reviews
interstellar_data = pd.DataFrame(reviews_list)

# Save DataFrame to a CSV file
interstellar_data.to_csv('interstellar_reviews.csv', index=False)
```
